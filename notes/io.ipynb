>  ## JSON read and write

```python
# read
import json
f = open('./assets/io.json').read()
params = json.loads(f)
print(params)
```

```python
# write
dict = {'property': 'content'}
with open('./assets/write.json', 'w') as f:
    json.dump(dict, f, indent=4, ensure_ascii=True)
```
